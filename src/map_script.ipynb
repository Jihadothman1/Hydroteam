```python
import dataretrieval.nwis as nwis
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import folium
import mplleaflet
from selenium import webdriver
```

```python
# calling specific data from USGS using NWIS function
# retrieves information from NWIS for the State of Utah




df = nwis.get_info(stateCd = 'UT')  
                             


#first row contains data available for the state of Utah in the NWIS database
#This includes the number of sites, the number of wells, the number of surface-water sites, the number
#of groundwater sites, the number of water-quality sites, and the number of other sites (such as atmospheric
#deposition sites, soil sites, and ecological sites).


#The second row contains information about the data availability for each year in the NWIS database for 
#the state of Utah. This includes the range of years for which data is available, the number of sites 
#with available data for each year, and the total number of measurements made for each year.


dfsites = df[0]
```

```python
#subsetting only the groundwater sites in the dataframe:

dfsitesGW = dfsites[dfsites["site_tp_cd"] == "GW"]
dfsitesGW
```

```python
#create a list of all wells mentioned in the NWIS of USGS


#Drop any duplicates in the sites to be able to have a list of the wells


dropppeddata = dfsitesGW.drop_duplicates(subset='site_no').copy()


unique_sites = dfsitesGW['site_no'].drop_duplicates()
unique_sites

#or use this one, but both give the same number of rows but it keeps the whole data:

siteslist = dfsitesGW.drop_duplicates(subset=['site_no'], keep='first')

if siteslist['dec_lat_va'].isna().any():
    print("There are NA values in the 'latitude' column.")

    
if siteslist['dec_long_va'].isna().any():
    print("There are NA values in the 'longitude' column.")

#NA_drop = df.dropna(subset=['dec_lat_va', 'dec_long_va'])


np.savetxt(r'C:\Users\jihad\Desktop\jihad\hydroteam\logan_wells\unique_sites.txt', unique_sites, fmt='%s')
np.savetxt(r'C:\Users\jihad\Desktop\jihad\hydroteam\logan_wells\siteslist.txt', siteslist, fmt='%s')
```

```python
#select only the first few rows to plot the location of the wells on the map. plotting 
#all of them would take a long time


new_df = siteslist.iloc[:20]
new_df
```

```python
map = folium.Map(location=[39.25, -111.45], zoom_start=8)
for index, row in new_df.iterrows():
    folium.Marker(location=[row['dec_lat_va'], row['dec_long_va']], 
                  popup=row['site_no']).add_to(map)
map.save('map.html')
    
    
    
# browser = webdriver.Firefox()
# browser.get(map._repr_html_())
# browser.save_screenshot(r'C:\Users\jihad\Desktop\jihad\hydroteam\logan_wells\map.png')
# browser.quit()

    
# fig, ax = plt.subplots()
# mplleaflet.show(fig=ax.figure)
# plt.savefig(r'C:\Users\jihad\Desktop\jihad\hydroteam\logan_wells\map.png')


#map.save(r'C:\Users\jihad\Desktop\jihad\hydroteam\logan_wells\map.html')
```
